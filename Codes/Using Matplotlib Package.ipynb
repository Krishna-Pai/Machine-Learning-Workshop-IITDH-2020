# Using Matplotlib Package

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a = np.arange(10)
b = a**2
```

```python
plt.plot(a,b,'r-o')
plt.xlabel('numbers')
plt.ylabel('squares')
```

```python
plt.plot(a,b,'g--')
plt.xlabel('numbers')
plt.ylabel('squares')
```

```python
plt.subplot(2,2,1)
plt.plot(a,b,'r-o')
plt.xlabel('numbers')
plt.ylabel('squares')

plt.subplot(2,2,4)
plt.plot(a,b,'g--')
plt.xlabel('numbers')
plt.ylabel('squares')
```
